```python
import pandas as pd

# Load your data
data = pd.read_csv('Data/combined_zpallagi.csv')

# Filter out rows where 'Std_Ded_Returns' or 'Itemized_Ded_Amt' are null
filtered_data = data.dropna(subset=['Itemized_Ded_Amt', 'Returns'])

# Aggregate data by state to calculate the total Std_Ded_Returns and Itemized_Ded_Amt per state
state_itemized_aggregated_adjusted = filtered_data.groupby('State').agg({
    'Itemized_Ded_Amt': 'sum',
    'Returns': 'sum'
}).reset_index()

# Calculate the percentage of Std_Ded_Returns relative to Itemized_Ded_Amt
state_itemized_aggregated_adjusted['Itemized_Ded_to_Returns_Percent'] = (state_itemized_aggregated_adjusted['Itemized_Ded_Amt'] / state_itemized_aggregated_adjusted['Returns']) * 100

# Output the results
print(state_itemized_aggregated_adjusted.head())
```

```python
# Calculate the percentage of Std_Ded_Returns relative to Returns
df['Percentage'] = (df['Std_Ded_Returns'] / df['Returns']) * 100

# Group data by 'State' and 'Year'
grouped = df.groupby(['State', 'year']).sum()

# Plotting the results
fig, axes = plt.subplots(nrows=len(grouped), ncols=2, figsize=(15, 6 * len(grouped)))
if axes.ndim == 1:  # If there's only one group, axes will be 1-dimensional
    axes = axes.reshape(-1, 2)

for ((state, year), data), (ax_pie, ax_bar) in zip(grouped.iterrows(), axes):
    # Pie chart for each group
    data['Percentage'].plot(kind='pie', autopct='%1.1f%%', ax=ax_pie)
    ax_pie.set_title(f'Pie Chart: {state} {year}')
    ax_pie.set_ylabel('')  # Remove y-label as it's unnecessary for pie charts

    # Bar chart for each group
    data['Percentage'].plot(kind='bar', ax=ax_bar, color='skyblue')
    ax_bar.set_title(f'Bar Chart: {state} {year}')
    ax_bar.set_xlabel('Percentage')
    ax_bar.set_ylabel('Frequency')
    ax_bar.set_xticklabels([f'{state} {year}'])

plt.tight_layout()
plt.show()
```

```python

# Defining the numbers
part = 44660
total = 2489580
remaining = total - part

# Data for the pie chart
labels_pie = ['Part (44,660)', 'Remaining (2,445,920)']
sizes_pie = [part, remaining]

# Plotting the pie chart
fig1, ax1 = plt.subplots()
ax1.pie(sizes_pie, labels=labels_pie, autopct='%1.1f%%', startangle=90)
ax1.axis('equal')  # Ensure that pie is drawn as a circle.
plt.title('Percentage Representation of 44,660 out of 2,489,580')

# Data for the bar chart
categories_bar = ['Part (44,660)', 'Remaining (2,445,920)']
values_bar = [part, remaining]

# Plotting the bar chart
fig2, ax2 = plt.subplots()
ax2.bar(categories_bar, values_bar, color=['blue', 'gray'])
ax2.set_ylabel('Amount')
ax2.set_title('Comparison of Part to Total')
ax2.set_xticks(range(len(categories_bar)))
ax2.set_xticklabels(categories_bar)

plt.show()
```
