```python

import os
import pandas as pd

# Get the current directory of the notebook
notebook_dir = os.getcwd()

# Navigate up one level to the parent directory
parent_dir = os.path.dirname(notebook_dir)

# Specify the relative path to the CSV file
relative_file_path = os.path.join(parent_dir, 'Data', 'combined_zpallagi.csv')

# Read the CSV file into a DataFrame
df = pd.read_csv(relative_file_path)

# Display the first few rows of the DataFrame to verify it's been loaded correctly
df.head()
```

```python
# Get the column titles
column_titles = df.columns

# Display the column titles
column_titles

```

```python
# Filter the DataFrame for the specified columns and rename it to usable_data_df
usable_data_df = df.loc[:, ['State', 'year', 'Itemized_Ded_Returns', 'Itemized_Ded_Amt', 'Charitable_Returns', 'Charitable_Amt']]

# Display the first few rows of the usable_data_df DataFrame
usable_data_df.head()
```

```python
# Set the option to display all rows
pd.set_option('display.max_rows', None)


```

```python
# Group the data by 'State' and 'year', then calculate the mean of 'Charitable_Amt'
avg_charitable_amt_by_state_year = usable_data_df.groupby(['State', 'year'])['Charitable_Amt'].sum()

# Convert the Series to a DataFrame
avg_charitable_amt_by_state_year_df = avg_charitable_amt_by_state_year.reset_index()

# Display the DataFrame
avg_charitable_amt_by_state_year_df
```

```python
import sqlite3

# Connect to the SQLite database (or create it if it doesn't exist)
conn = sqlite3.connect('heatmap_data.db')

# Write the DataFrame to a SQLite database table called "heatmap_data"
avg_charitable_amt_by_state_year_df.to_sql('heatmap_data', conn, index=False, if_exists='replace')

# Close the connection
conn.close()
```

```python
# Get the top 5 values in charitable amounts
top_charitable_amounts = avg_charitable_amt_by_state_year_df.nlargest(5, 'Charitable_Amt')

# Get the lowest 5 values in charitable amounts
lowest_charitable_amounts = avg_charitable_amt_by_state_year_df.nsmallest(5, 'Charitable_Amt')

# Display the top and lowest values
print("Top 5 values in charitable amounts:")
print(top_charitable_amounts)

print("\nLowest 5 values in charitable amounts:")
print(lowest_charitable_amounts)
```

```python
import matplotlib.pyplot as plt

# Plotting the top 5 values
plt.figure(figsize=(10, 6))
plt.bar(range(len(top_charitable_amounts)), top_charitable_amounts['Charitable_Amt'], color='blue')
plt.title('Top 5 States by Charitable Amounts')
plt.xlabel('State and Year')
plt.ylabel('Charitable Amount')
plt.xticks(range(len(top_charitable_amounts)), [f"{state}, {year}" for state, year in zip(top_charitable_amounts['State'], top_charitable_amounts['year'])], rotation=45)
plt.tight_layout()
plt.show()

# Plotting the lowest 5 values
plt.figure(figsize=(10, 6))
plt.bar(range(len(lowest_charitable_amounts)), lowest_charitable_amounts['Charitable_Amt'], color='red')
plt.title('Lowest 5 States by Charitable Amounts')
plt.xlabel('State and Year')
plt.ylabel('Charitable Amount')
plt.xticks(range(len(lowest_charitable_amounts)), [f"{state}, {year}" for state, year in zip(lowest_charitable_amounts['State'], lowest_charitable_amounts['year'])], rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Combine the top and lowest values into a single DataFrame
combined_data = pd.concat([top_charitable_amounts, lowest_charitable_amounts])

# Sort the combined DataFrame by 'Charitable_Amt'
combined_data = combined_data.sort_values(by='Charitable_Amt')

# Plotting the combined data
plt.figure(figsize=(10, 6))
plt.bar(range(len(combined_data)), combined_data['Charitable_Amt'], color=['blue'] * 5 + ['red'] * 5)
plt.title('Top 5 and Lowest 5 States by Charitable Amounts')
plt.xlabel('State and Year')
plt.ylabel('Charitable Amount')
plt.xticks(range(len(combined_data)), [f"{state}, {year}" for state, year in zip(combined_data['State'], combined_data['year'])], rotation=45)
plt.tight_layout()
plt.show()
```

```python
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect('heatmap_data.db')

# Create a cursor object to execute SQL queries
cursor = conn.cursor()

# Execute a SELECT query to retrieve data from the table
cursor.execute('SELECT * FROM heatmap_data')

# Get the column names from the cursor description
columns = [column[0] for column in cursor.description]

# Print the column names
print("Column Names:", columns)

# Fetch all rows from the result set
rows = cursor.fetchall()

# Map the column names to the original DataFrame column names
mapped_rows = [{columns[i]: row[i] for i in range(len(columns))} for row in rows]

# Print the mapped rows
for row in mapped_rows:
    print(row)

# Close the cursor and the connection
cursor.close()
conn.close()
```
